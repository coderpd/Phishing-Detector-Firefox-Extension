```python
from pandas import read_csv, DataFrame
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_selection import RFE
```

```python
X = read_csv('dataset/Dataset.csv', header=None)
X.head()
```

```python
shape = X.shape
print("Number of training examples: ", shape[0])
print("Number of features: ", shape[1])
```

```python
X.columns[X.isnull().any()]
```

```python
Y = read_csv('dataset/Target_Labels.csv', header=None)
Y.head()
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
display(
    train_X.shape,
    train_Y.shape,
    test_X.shape,
    test_Y.shape    
)
```

```python
svc = SVC(kernel='linear', C=1)
rfe = RFE(estimator=svc, n_features_to_select=train_X.columns.size)
```

```python
rfe.fit(train_X, train_Y)
```

```python
predict = rfe.predict(test_X)
```

```python
accuracy_score(predict, test_Y)
```

```python
print(classification_report(predict, test_Y))
```

```python
print(confusion_matrix(predict, test_Y))
```
